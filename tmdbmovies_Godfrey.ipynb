```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Load the dataset
df = pd.read_csv('tmdb.movies.csv')

## Display the first few rows of the dataset
print(df.head())

## Display the columns of the dataset
print(df.columns)

## Display the data types of the columns
print(df.dtypes)

## Display the shape of the dataset
print(df.shape)
```

```python
#Cleaning the dataset.
#Check for missing values
print(df.isnull().sum())
```

```python
#Drop unnecessary columns.
df.drop(['id', 'genre_ids', 'original_language', 'original_title', 'release_date'], axis=1, inplace=True)
```

```python
#Check top ten movies with the highest popularity.
top_10_popularity = df.nlargest(10, 'popularity')
print(top_10_popularity[['title', 'popularity']])

#Plotting the top 10 movies with the highest popularity.
plt.figure(figsize=(10, 6))
sns.barplot(x='popularity', y='title', data=top_10_popularity, palette='viridis')
plt.title('Top 10 Movies by Popularity')
plt.xlabel('Popularity')
plt.ylabel('Movie Title')
plt.show()
```

```python
#Check top ten movies with the highest votes
top_10_votes = df.nlargest(10, 'vote_count')
print(top_10_votes[['title', 'vote_count']])

#Plotting the top 10 movies with the highest votes.
plt.figure(figsize=(10, 6))
sns.barplot(x='vote_count', y='title', data=top_10_votes, palette='viridis')
plt.title('Top 10 Movies by Vote Count')
plt.xlabel('Vote Count')
plt.ylabel('Movie Title')
plt.show()
```
